```python


def plot_histogram(df, feature):
    plt.figure(figsize=(9, 4))
    sns.histplot(df[feature], bins=50, kde=True, color='skyblue')
    plt.xlabel(feature)
    plt.ylabel("Frequency")
    plt.title(f"Distribution of {feature}")
    plt.show()

def plot_variables(df, variables):
    for variable in variables:
        plot_histogram(df, variable)

all_variables = ["Education", "Marital_Status","Income","Complain", "Year_Birth", "Kidhome", "Teenhome"]

plot_variables(df, all_variables)


```

```python
from sklearn.cluster import KMeans

# List to store the Within-Cluster-Sum-of-Squares (WCSS) values for different values of k
wcss = []

# Iterate through different values of k (number of clusters)
for k in range(1, 15):
    # Create a KMeans clustering model with the current value of k
    kmeans = KMeans(n_clusters=k)
    kmeans.fit(df)
    wcss.append(kmeans.inertia_)

# Plot the WCSS values against the number of clusters (k)
plt.plot(range(1, 15), wcss)
plt.xlabel("Number of Clusters (k)")
plt.xticks(range(1, 15, 1))
plt.ylabel("Within-Cluster-Sum-of-Squares (WCSS) Value")
plt.title("Elbow Method for Optimal k Selection")
plt.show()
```

```python
from sklearn.decomposition import PCA
import pandas as pd

# Create a PCA (Principal Component Analysis) object with 3 components and whiten the data
pca = PCA(n_components=3, whiten=True)

# Fit the PCA model to the data
pca.fit(df)

# Transform the original data into the principal components
data_pca = pca.transform(df)

# Create a DataFrame to represent the principal components with column names
pca_df = pd.DataFrame(data_pca, columns=['PC1', 'PC2', 'PC3'])

# Print the explained variance ratio which indicates the contribution of each original feature to the principal components
print("Explained Variance Ratio:")
print(pca.explained_variance_ratio_)

# You can also print the loadings to understand which original features contribute most to each principal component
print("Loadings (Eigenvectors):")
print(pca.components_)
```

```python
data_pca
```

```python
kmeans2 = KMeans(n_clusters=3)

clusters = kmeans2.fit_predict(data_pca)
```

```python
from mpl_toolkits.mplot3d import Axes3D
colors = plt.cm.get_cmap('tab10', 3)

# Plotting the results in 3D
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

for cluster_num in range(3):
    cluster_indices = np.where(clusters == cluster_num)[0]  # Get indices of data points in this cluster
    ax.scatter(data_pca[cluster_indices, 0], 
               data_pca[cluster_indices, 1], 
               data_pca[cluster_indices, 2], 
               label=f'Cluster {cluster_num + 1}', 
               color=colors(cluster_num))

ax.set_title('PCA with Cluster Coloring in 3D')
ax.set_xlabel('PCA Component 1')
ax.set_ylabel('PCA Component 2')
ax.set_zlabel('PCA Component 3')
ax.legend()
plt.show()
```
